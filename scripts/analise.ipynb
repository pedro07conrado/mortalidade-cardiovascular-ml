```python
import pandas as pd

# 1. Carregar o arquivo salvo
df = pd.read_csv("populacao_municipios_ibge.csv")

# 2. Ver o total de munic√≠pios √öNICOS (sem repetir por ano)
qtd_municipios = df['codmun'].nunique()
print(f"üè† Total de munic√≠pios √∫nicos encontrados: {qtd_municipios}")

# 3. Ver quantos munic√≠pios tem em CADA ANO
print("\nüìÖ Munic√≠pios por ano:")
print(df['ano'].value_counts().sort_index())

# 4. Verificar se h√° duplicatas (erros)
duplicados = df.duplicated(subset=['codmun', 'ano']).sum()
print(f"\n‚ö†Ô∏è Linhas duplicadas (munic√≠pio + ano repetido): {duplicados}")
```
